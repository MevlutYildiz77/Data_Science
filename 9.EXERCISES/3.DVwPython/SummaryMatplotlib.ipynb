# Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

a=plt.plot(x, y)
a
plt.show()
```

```python
plt.plot(x, y)        # plot x and y using default line style and color
```

```python
plt.plot(x, y, 'bo')  # plot x and y using blue circle markers
```

```python
plt.plot(y)           # plot y using x as index array 0..N-1
```

```python
plt.plot(y, 'r+')     # ditto, but with red plusses
```

```python
plt.plot(x, y, 'go--', linewidth=2, markersize=12)
```

```python
plt.plot(x, y, color='green', marker='o', linestyle='dashed',
...      linewidth=2, markersize=12)
```

```python
plt.plot('xlabel', 'ylabel', data=a)   # 
```

```python
plt.plot(x[0], y[1:])
```

```python
 plt.plot(x, y, 'bo')
```

# Matplotlib Markers

```python
plt.plot(y, marker='o')  #markers can be as bottom markers ; 
'o'	Circle	
'*'	Star	
'.'	Point	
','	Pixel	
'x'	X	
'X'	X (filled)	
'+'	Plus	
'P'	Plus (filled)	
's'	Square	
'D'	Diamond	
'd'	Diamond (thin)	
'p'	Pentagon	
'H'	Hexagon	
'h'	Hexagon	
'v'	Triangle Down	
'^'	Triangle Up	
'<'	Triangle Left	
'>'	Triangle Right	
'1'	Tri Down	
'2'	Tri Up	
'3'	Tri Left	
'4'	Tri Right	
'|'	Vline	
'_'	Hline
```

# Line Reference"

```python
plt.plot(y, 'o:r')  #'r' is a line here. Lines can be as below:
plt.show()

'-'	Solid line	
':'	Dotted line	
'--'	Dashed line	
'-.'	Dashed/dotted line
```

# Color Reference

```python
plt.plot(y, 'o:r')  #'o' is a color here. Colors can be as below:
plt.show()

#'r'	Red	
#'g'	Green	
#'b'	Blue	
#'c'	Cyan	
#'m'	Magenta	
#'y'	Yellow	
#'k'	Black	
#'w'	White
```
